```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import os

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# load all data
X = []
y = []
y_f0 = []
max_seq_size = 0
data_path = "data-processed"
data_points = [point for point in os.listdir(data_path) if point != '.DS_Store']
for dir in data_points:
    data_point_path = os.path.join(data_path, dir)
    for file_name in os.listdir(data_point_path):
        #if file_name.endswith('.npy'):
        if file_name.endswith('pre-processed.npy'):
            filepath = os.path.join(data_point_path, file_name)
            file = np.load(filepath)
            if file.shape[1] > max_seq_size:
                max_seq_size = file.shape[1]
            if file_name.split('-')[3] == 'left':
                #print(file_name)
                X.append(file)
                if file_name.split('-')[1] == 'static':
                    y.append(0) # static
                if file_name.split('-')[1] == 'vertical':
                    y.append(1) # vertical
                if file_name.split('-')[1] == 'horizontal':
                    y.append(2) # horizontal
                if file_name.split('-')[1] == 'circular':
                    y.append(3) # circular
        if file_name.endswith('f0.npy'):
            filepath = os.path.join(data_point_path, file_name)
            file = np.load(filepath)
            y_f0.append(file)

print(f'Max sequence size: {max_seq_size}')
print(f'X size: {len(X)}')
print(f'y size: {len(y)}')
print(f'y_f0 size: {len(y_f0)}')
```

# Regression of vibrato depth and frequency from sensor data

```python
import librosa
sr = 44100
cent_static = y_f0[0]
avg_freqs = []
vibrato_depths = []
X_filtered = []
y_filtered = []
threshold_keep = 200
for i in range(len(y_f0)):
    cent_static = y_f0[i]
    if cent_static.shape[0] > threshold_keep:
        #print(cent_static.shape)
        cent_static_delta = librosa.feature.delta(cent_static)
        
        cent_static_delta = np.array(cent_static_delta)
        cent_static_delta_filtered = cent_static_delta[(np.abs(stats.zscore(cent_static_delta)) < 1.5)]
        vibrato_depth = cent_static_delta_filtered.max() - cent_static_delta_filtered.min()

        #fig, ax = plt.subplots(figsize=(8,1))
        #ax.plot(cent_static_delta_filtered)
        
        zero_crossings = np.where(np.diff(np.sign(cent_static_delta_filtered)))[0]
        zero_crossings_times = librosa.frames_to_time(zero_crossings, sr=sr)
        periods = [zero_crossings_times[i+1] - zero_crossings_times[i] for i in range(zero_crossings_times.shape[0]-1)]
        avg_period = np.array(periods).mean()
        std_period = np.array(periods).std()
        avg_freq = 1 / avg_period
    
        X_filtered.append(X[i])
        y_filtered.append(y[i])
        vibrato_depths.append(vibrato_depth)
        avg_freqs.append(avg_freq)
        
        #print(f'vibrato depth: {vibrato_depth:.3f}')
        #print(f'vibrato frequency: {avg_freq:.3f}')
```

```python
X_filtered_mean = [x.mean(axis=1) for x in X_filtered]
X_filtered_std = [x.std(axis=1) for x in X_filtered]
X_filtered_mean = np.array(X_filtered_mean)
X_filtered_std = np.array(X_filtered_std)
y_filtered = np.array(y_filtered)
vibrato_depths = np.array(vibrato_depths)
avg_freqs = np.array(avg_freqs)

X_filtered_mean_std = np.hstack((X_filtered_mean,X_filtered_std))
X_filtered_mean_std.shape
```

```python
vibrato_depths0 = [vibrato_depths[i] for i in range(len(vibrato_depths)) if y_filtered[i] == 0]
vibrato_depths1 = [vibrato_depths[i] for i in range(len(vibrato_depths)) if y_filtered[i] == 1]
vibrato_depths2 = [vibrato_depths[i] for i in range(len(vibrato_depths)) if y_filtered[i] == 2]
vibrato_depths3 = [vibrato_depths[i] for i in range(len(vibrato_depths)) if y_filtered[i] == 3]

vibrato_depths0 = np.array(vibrato_depths0)
vibrato_depths0_filtered = vibrato_depths0[(np.abs(stats.zscore(vibrato_depths0)) < 3)]
vibrato_depths1 = np.array(vibrato_depths1)
vibrato_depths1_filtered = vibrato_depths1[(np.abs(stats.zscore(vibrato_depths1)) < 3)]
vibrato_depths2 = np.array(vibrato_depths2)
vibrato_depths2_filtered = vibrato_depths2[(np.abs(stats.zscore(vibrato_depths2)) < 3)]
vibrato_depths3 = np.array(vibrato_depths3)
vibrato_depths3_filtered = vibrato_depths3[(np.abs(stats.zscore(vibrato_depths3)) < 3)]

print(np.array(vibrato_depths0_filtered).mean())
print(np.array(vibrato_depths1_filtered).mean())
print(np.array(vibrato_depths2_filtered).mean())
print(np.array(vibrato_depths3_filtered).mean())
print(np.array(vibrato_depths0_filtered).std())
print(np.array(vibrato_depths1_filtered).std())
print(np.array(vibrato_depths2_filtered).std())
print(np.array(vibrato_depths3_filtered).std())
```

```python
avg_freqs0 = [avg_freqs[i] for i in range(len(avg_freqs)) if y_filtered[i] == 0]
avg_freqs1 = [avg_freqs[i] for i in range(len(avg_freqs)) if y_filtered[i] == 1]
avg_freqs2 = [avg_freqs[i] for i in range(len(avg_freqs)) if y_filtered[i] == 2]
avg_freqs3 = [avg_freqs[i] for i in range(len(avg_freqs)) if y_filtered[i] == 3]

avg_freqs0 = np.array(avg_freqs0)
avg_freqs0_filtered = avg_freqs0[(np.abs(stats.zscore(avg_freqs0)) < 3)]
avg_freqs1 = np.array(avg_freqs1)
avg_freqs1_filtered = avg_freqs1[(np.abs(stats.zscore(avg_freqs1)) < 3)]
avg_freqs2 = np.array(avg_freqs2)
avg_freqs2_filtered = avg_freqs2[(np.abs(stats.zscore(avg_freqs2)) < 3)]
avg_freqs3 = np.array(avg_freqs3)
avg_freqs3_filtered = avg_freqs3[(np.abs(stats.zscore(avg_freqs3)) < 3)]

print(np.array(avg_freqs0_filtered).mean())
print(np.array(avg_freqs1_filtered).mean())
print(np.array(avg_freqs2_filtered).mean())
print(np.array(avg_freqs3_filtered).mean())
print(np.array(avg_freqs0_filtered).std())
print(np.array(avg_freqs1_filtered).std())
print(np.array(avg_freqs2_filtered).std())
print(np.array(avg_freqs3_filtered).std())
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

X_filtered_mean_std = X_filtered_mean_std[~np.isnan(avg_freqs),:]
y_filtered = y_filtered[~np.isnan(avg_freqs)]
avg_freqs = avg_freqs[~np.isnan(avg_freqs)]

X_filtered_mean_std = X_filtered_mean_std[(np.abs(stats.zscore(avg_freqs)) < 3)]
y_filtered = y_filtered[(np.abs(stats.zscore(avg_freqs)) < 3)]
avg_freqs = avg_freqs[(np.abs(stats.zscore(avg_freqs)) < 3)]

N = 1 # polynomial degree
poly = PolynomialFeatures(N)
X_poly = poly.fit_transform(X_filtered_mean_std)

# model hyperparameters
N = 2
y = avg_freqs.reshape(-1,1)
test_size = 0.2
# train-test split
X_train, X_test, y_train, y_test = train_test_split(X_filtered_mean_std, y, test_size=test_size, random_state=666, shuffle=True)

# Normalize the data
X_scaler = preprocessing.MinMaxScaler()
y_scaler = preprocessing.MinMaxScaler()

# Test values have to be normalized with the training mean and std
y_scaler.fit(y_train)
y_train = y_scaler.transform(y_train)
y_test = y_scaler.transform(y_test)
X_scaler.fit(X_train)
X_train = X_scaler.transform(X_train)
X_test = X_scaler.transform(X_test)

reg = LinearRegression().fit(X_train, y_train)
train_score = reg.score(X_train, y_train)
test_score = reg.score(X_test, y_test)
print(f'LR degree 1 train score: {train_score:.6f}')
print(f'LR degree 1 test score: {test_score:.6f}')
print()

from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=4, random_state=0).fit(X_train, y_train)
train_score = regr.score(X_train, y_train)
test_score = regr.score(X_test, y_test)
print(f'RF degree 1 train score: {train_score:.6f}')
print(f'RF degree 1 test score: {test_score:.6f}')
print()

from sklearn.svm import SVR
regr = SVR(C=1.0, epsilon=0.2).fit(X_train, y_train)
train_score = regr.score(X_train, y_train)
test_score = regr.score(X_test, y_test)
print(f'SVR degree 1 train score: {train_score:.6f}')
print(f'SVR degree 1 test score: {test_score:.6f}')
print()
```
